```python
import potentials
```

```python
db = potentials.Database()
```

```python
pots, pots_df = db.get_potentials(return_df=True)
```

```python
def multicites(series):
    return len(series.citations) > 1 
```

```python
for potential in pots[pots_df.apply(multicites, axis=1)]:
    potential.html(render=True)
```
