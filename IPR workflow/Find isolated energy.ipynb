```python
import numpy as np

# https://github.com/usnistgov/iprPy
import iprPy
print('iprPy version', iprPy.__version__)
```

```python
database = iprPy.load_database('master')
results = database.get_records_df(style='calculation_diatom_scan', full=True, flat=False)
```

```python
for i in range(len(results)):
#for i in range(1):
    series = results.iloc[i]
    matchr = []
    matche = []
    for j in range(len(series.diatom_plot)-1):
        rs = series.diatom_plot.r.values
        energies = series.diatom_plot.energy.values
        if energies[j] == energies[j+1] and energies[j] not in matche:
            matchr.append(rs[j])
            matche.append(energies[j])
    matches = {}
    for j in range(len(matchr)):
        if matchr[j] != 0.02:
            matches[matchr[j]] = matche[j]
    print(matches)    
```
